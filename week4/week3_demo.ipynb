# 2. 國立？所；私立？所

step1: 建立空字串 (lsit=[])

step2: 建立for迴圈，配合if else條件，取出國立與私立

step3: 新增「公私立」column在DataFrame中 (df['公私立'])

step4: 篩選掉重複的學校代碼資料 (drop_duplicates)

step5: 計算公立與私立學校的筆數，並印出結果 (value_counts)

```python
type_list = [] # 建立空字串

for i in df['學校名稱']:
    if ('國立' in i) or ('市立' in i):
        type_list.append('國立')
    else:
        type_list.append('私立')

df['公私立'] = type_list # 將 Dataframe 新增「公私立」column

df2 = df.drop_duplicates(['學校代碼'],keep='first')
count = len(df2['公私立'])
count1 = len(df2[df2['公私立'] == '國立'])
count2 = len(df2[df2['公私立'] == '私立'])
print(df2['公私立'])
print(f"本資料集共收集了 {count} 所學校，其中公立：{count1} 所；私立：{count2} 所。")
```

# 5. 北中南東區分別有？所

中部區域：包括臺中市、苗栗縣、彰化縣、南投縣及雲林縣。

南部區域：包括高雄市、臺南市、嘉義市、嘉義縣、屏東縣及澎湖縣。

東部區域：包括花蓮縣及臺東縣。

# 自選1各大學性別比為何?


```python
dfA = df.drop_duplicates(['男生計','女生計','學校名稱'],keep='first')

school_list = list(pd.unique(df['學校名稱']))

for school in school_list:
  dfSchool = dfA[dfA['學校名稱']==school]
  male = dfSchool['男生計']
  female = dfSchool['女生計']
  maleSum = male.mean(axis=0)*len(male)
  femaleSum = female.mean(axis=0)*len(female)
  ratioM = maleSum/femaleSum*100
  #print(male)
  print(f"{school}的男女比為 {ratioM.round(1)} 。")
```

# 自選2各校延畢生數量統計

```python
dfB = df.drop_duplicates(['延修生男','延修生女','學校名稱'],keep='first')

#school_list = list(pd.unique(df['學校名稱']))

for school in school_list:
  dfSchool = dfB[dfB['學校名稱']==school]
  maleY = dfSchool['延修生男']
  femaleY = dfSchool['延修生女']
  maleYSum = maleY.mean(axis=0)*len(maleY)
  femaleYSum = femaleY.mean(axis=0)*len(femaleY)
  sumY = maleYSum+femaleYSum
  #print(male)
  print(f"{school}的延畢生數量為 {sumY.round(1)} 。")
```

# 自選3各縣市學生數統計

```python
dfC = df.drop_duplicates(['總計','縣市名稱'],keep='first')

city_list = list(pd.unique(df['縣市名稱']))

for city in city_list:
  dfCity = dfC[dfC['縣市名稱']==city]
  stu = dfCity['總計']
  stuSum = stu.mean(axis=0)*len(stu)
  #print(male)
  print(f"{city}的學生數為 {stuSum.round(1)} 。")
```
